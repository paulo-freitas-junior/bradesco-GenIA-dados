```python
import os
from dotenv import load_dotenv
from scipy.io.wavfile import write
import numpy as np
import openai
```

# 1. Carregar chave da API do .env

```python
load_dotenv()
API_KEY = os.getenv("OPENAI_API_KEY")
if not API_KEY:
    raise ValueError("Chave da API OpenAI não encontrada. Verifique no arquivo .env se a mesma econtra-se configurada.")
openai.api_key = API_KEY
```

# 2. Definir pasta de saída para os áudios

```python
OUTPUT_DIR = "./data/audio"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def salvar_audio(record_output: str, taxa_amostragem: int, dados: np.ndarray):
    """Salva um array numpy como arquivo .wav na pasta audios"""
    caminho = os.path.join(OUTPUT_DIR, f"{record_output}.wav")
    write(caminho, taxa_amostragem, dados)
    print(f"Áudio salvo em: {caminho}")
```

# 3. Exemplo de uso com dados fictícios
No código real, 'dados' virá do Whisper ou outra lib

```python
fs = 16000  # taxa de amostragem
dados = np.random.randint(-32768, 32767, fs*2, dtype=np.int16)  # 2 segundos de áudio fake

salvar_audio("teste", fs, dados)
```
